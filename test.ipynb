```python
import pandas as pd
import networkx as nx

#Visualize identified entities
doc = nlp(text)
displacy.render(doc, style="ent", jupyter=True)
```

```python
# Get named entity list per sentence

sent_entity_df = []

for sent in doc.sents:
    entity_list = [ent.text for ent in sent.ents]
    sent_entity_df.append({"sentence": sent, "entities": entity_list})
    
sent_entity_df = pd.DataFrame(sent_entity_df)
```

```python
sent_entity_df
```

```python
sent_entity_df_filtered = sent_entity_df[sent_entity_df['entities'].map(len) > 0]
sent_entity_df_filtered
```

```python
window_size = 15
relationships = []

for i in range(sent_entity_df_filtered.index[-1]):
    end_i = min(i + 5, sent_entity_df_filtered.index[-1])
    ent_list = sum(sent_entity_df_filtered.loc[i: end_i].entities, [])
    
    ent_unique = [ent_list[i] for i in range(len(ent_list)) if (i==0) or ent_list[i] != ent_list[i-1]]
    
    if len(ent_unique) > 0:
        for idx, a in enumerate(ent_unique[:-1]):
            b = ent_unique[idx+1]
            relationships.append({"source": a, "target": b})

relationships_df = pd.DataFrame(relationships)

pd.set_option('display.max_rows', None)
relationships_df
```

```python
relationships_df = pd.DataFrame(np.sort(relationships_df.values, axis=1), columns= relationships_df.columns)
relationships_df
```

```python
relationships_df["value"] = 1
relationships_df = relationships_df.groupby(["source", "target"], sort=False, as_index=False).sum()

relationships_df
```

```python
G = nx.from_pandas_edgelist(relationships_df, source="source", target="target", edge_attr="value")

pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels=True)
```

```python
from pyvis.network import Network
net = Network(notebook=True)

net.from_nx(G)
net.show("graph_test.html")
```
